# Upper

## After Cropping

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load the dataset
ds = xr.open_dataset('/pfs/lustrep1/scratch/project_462000472/akhtar/climate_modeling/pangu-data/cropped_with_pm2.5_model/upper/upper_20180101.nc')
print(ds)

# Extract the first time point
first_time = ds.isel(time=0)

# Set up the plotting figure
fig, axs = plt.subplots(3, 2, figsize=(15, 15), subplot_kw={'projection': ccrs.PlateCarree()})

# Define the variables and their titles, along with color maps
variables = ['z', 'q', 't', 'u', 'v']
titles = ['Height (z)', 'Specific Humidity (q)', 'Temperature (t)', 'U-Component (u)', 'V-Component (v)']
colormaps = ['coolwarm', 'coolwarm', 'coolwarm', 'seismic', 'seismic']

# Loop over the variables and plot each one
for i, var in enumerate(variables):
    ax = axs[i // 2, i % 2]  # Adjust subplot position
    data = first_time[var].isel(level=0)  # Plot at the first level
    
    # Plot the data with the specified colormap
    im = data.plot(ax=ax, cmap=colormaps[i], add_colorbar=False)

    # Add continent lines and labels
    ax.coastlines(resolution='110m', color='black')
    ax.add_feature(cfeature.BORDERS, linestyle=':', color='black')
    ax.set_title(titles[i])

# Remove the last empty subplot
fig.delaxes(axs[2, 1])

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

# PM2.5

### Before Regridding

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Path to your .nc file
file_path = "/pfs/lustrep1/scratch/project_462000472/akhtar/climate_modeling/pangu-data/data_prep/pm25/original_075/2018/cams_201801.nc"
ds = xr.open_dataset(file_path)
print(ds)

# Select the first time step for visualization
pm2p5_data = ds.pm2p5.isel(time=0)

# Create the plot
plt.figure(figsize=(12, 6))

# Set up a Cartopy projection for the map
ax = plt.axes(projection=ccrs.PlateCarree())
heatmap = plt.pcolormesh(
    pm2p5_data.longitude, 
    pm2p5_data.latitude, 
    pm2p5_data, 
    transform=ccrs.PlateCarree(), 
    shading="auto", 
    cmap="viridis"
)

# Add a colorbar
plt.colorbar(heatmap, label="PM2.5 Concentration (units)")

# Add features: land, coastlines, and borders
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linestyle=':', linewidth=0.5, edgecolor='black')

# Add title and axis labels
plt.title("PM2.5 Concentration at First Time Step")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.show()
```

```python
max_value = pm2p5_data.max().item()  # Convert to a Python scalar
min_value = pm2p5_data.min().item()  # Convert to a Python scalar

print(f"Maximum PM2.5 concentration: {max_value}")
print(f"Minimum PM2.5 concentration: {min_value}")
```
